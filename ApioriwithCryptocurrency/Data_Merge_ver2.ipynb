```python
import pandas as pd
from apyori import apriori
```

## Caculate percentege change in coin value

```python
def percentage_change(col1,col2):
    return ((col2 - col1) / col1) * 100
```

## Transform data in excel file

```python
def turnText(df1):
  perc= [] 
  for i in range(0, len(df1)):
    val = percentage_change(df1.iloc[i]['Open'], df1.iloc[i]['Close'])
    if 0<= val <= 1:
      perc.append("_No_Change")
    elif 1 < val <= 5:
      perc.append("_In_5")
    elif 5 < val <=10:
      perc.append("_In_10")
    elif 10 < val <=20:
      perc.append("_In_20")
    elif 20 < val:
      perc.append("_Above_20")
    elif -5 <= val < 0:
      perc.append("_De_5")
    elif -10 <= val < -5:
      perc.append("_De_10")
    elif -20 <= val < -10:
      perc.append("_De_20")
    else:
      perc.append("_Below_20")
    
  return perc
```

```python
def turnData(df1):
  perc= [] 
  for i in range(0, len(df1)):
    val = percentage_change(df1.iloc[i]['Open'], df1.iloc[i]['Close'])
    if 0<= val <= 1:
      perc.append("_No Change")
    elif 1 < val:
      perc.append("_Increase")
    else:
      perc.append("_Decrease")
  return perc
```

## Choose 1 code cell below to run

### This code cell to transfrom data to Increase, Decrease, Nochange 

```python
def readFile(filename):
    df = pd.read_csv(filename + ".csv", usecols=['Date', 'Open', 'Close'], index_col = 'Date', parse_dates = True)
    #df.columns = [filename + '_O', filename + '_C']
    df[filename] = [filename + s  for s in turnData(df)]
    df = df.drop(columns=['Open', 'Close'])
    return df
```

### This code cell to transfrom data to percentage version

```python
def readFile(filename):
    df = pd.read_csv(filename + ".csv", usecols=['Date', 'Open', 'Close'], index_col = 'Date', parse_dates = True)
    #df.columns = [filename + '_O', filename + '_C']
    df[filename] = [filename + s  for s in turnText(df)]
    df = df.drop(columns=['Open', 'Close'])
    return df
```

## Read file

```python
file_csv = ["coin_Aave", "coin_BinanceCoin", "coin_Bitcoin", "coin_Cardano", "coin_ChainLink", "coin_Cosmos", "coin_CryptocomCoin"
            , "coin_Dogecoin", "coin_EOS", "coin_Ethereum", "coin_Iota", "coin_Litecoin", "coin_Monero", "coin_NEM", "coin_Polkadot"
            , "coin_Solana", "coin_Stellar", "coin_Tether", "coin_Tron", "coin_Uniswap", "coin_USDCoin", "coin_WrappedBitcoin", "coin_XRP"]
```

```python
dfs = []
for names in file_csv:
    dfs.append(readFile(names))
```

```python
from functools import reduce
dataset = reduce(lambda left,right: pd.merge(left,right,on='Date', how='outer'), dfs)
```

```python
dataset = dataset.sort_values("Date")
```

```python
dataset
```

## Run Apiori

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0]) for result in results]
    rhs         = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
```

```python
transactions = []
resultsinDataFrame = []

for i in range(0, len(dataset)):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 23) if str(dataset.values[i,j])!='nan' ])
rules = apriori(transactions = transactions, min_support = 0.08, min_confidence = 0.7, min_lift = 2, max_length = 3)
results = list(rules)
resultsinDataFrame.append(pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift']))
```

### Shift day by 1, repeat with all coin

```python

for names in file_csv:
    dataset_shift = dataset.copy()
    dataset_shift[dataset_shift.columns.difference([names])] = dataset_shift.loc[:, dataset_shift.columns != names].shift(-1)
    transactions_shift = []
    for i in range(0, len(dataset_shift)):
      transactions_shift.append([str(dataset_shift.values[i,j]) for j in range(0, 23) if str(dataset_shift.values[i,j])!='nan' ])
    
    rules = apriori(transactions = transactions_shift, min_support = 0.1, min_confidence = 0.7, min_lift = 2, max_length = 3)
    results = list(rules)
    resultsinDataFrame.append(pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift']))
    
```

## Remove duplicate rules

```python
from functools import reduce
dataFr= reduce(lambda left,right: pd.merge(left,right, on=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'], how='outer'), resultsinDataFrame)
```

```python
dataFr = dataFr.drop_duplicates(subset=['Left Hand Side', 'Right Hand Side']).sort_values("Lift", ascending = False)
#dataFr.drop_duplicates(subset=['Left Hand Side', 'Right Hand Side'])
```
